```python
import numpy as np
import matplotlib.pyplot as plt
```

#### I used this nb to find the appropriate scoring function for the guesser game

```python
max_points = 100

def score(dist, max_dist, k=1e-3):
    return round(max_points * (np.exp(-dist * k) - np.exp(-max_dist * k)) / (1 - np.exp(-max_dist * k)))

# map -> (max_dist, k)
maps = {
    'World': (np.sqrt(149_000_000), 5e-4),
    'Europe': (np.sqrt(10_180_000), 2e-3),
    'Asia': (np.sqrt(44_579_000), 5e-4),
    'Africa': (np.sqrt(30_200_000), 5e-4),
    'North America': (np.sqrt(24_290_000), 5e-4),
    'South America': (np.sqrt(17_840_000), 5e-4),
    'Oceania': (np.sqrt(8_520_000), 7.5e-4),
}


def score_range(range, max_dist, k=1e-3):
    return [score(x, max_dist, k) for x in range]


def plot_score_function(map_name, map_max_dist, k=1e-3):
    dists = np.linspace(0, max_dist, 10_000)
    plt.figure(figsize=(12,6))
    plt.plot(dists, score_range(dists, max_dist, k))
    plt.xlabel('Distance (km)')
    plt.ylabel('Score')
    plt.title(f'Map: {map_name} - Score Function')
    plt.grid(True,alpha=0.3)
    plt.show()



for map_name, (max_dist, k) in maps.items():
    plot_score_function(map_name, max_dist, k)
```
